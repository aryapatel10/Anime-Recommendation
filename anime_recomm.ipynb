```python
from sklearn.feature_extraction.text import CountVectorizer#CountVectorizer is class
cv=CountVectorizer(max_features=2000,stop_words='english')#create object cv
#stop_words->are,and,on,in etc
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    lst=[]
    for i in text.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
stem('Kimi no Na wa.Drama, Romance, School, Supernatural')
```

```python
ps.stem('dancing','danced')
```

```python
anime['tags']=anime['tags'].apply(stem)
```

```python
#we can count distance between two vector using euclid and cosine so we will go for cosine bcz it is more efficience for our case
#dist=1/similarity
#in higher dimension euclidean distance is not realible measure so we will use cosine distance it is theeta between two vector(movie)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#similarity=1/dist
#similarity between 1 to 0 if 1 it mean 100% match if 0 mean no matching
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recom(anime_name):
    anime_index=anime[anime['name']==anime_name].index[0]
    dist=similarity[anime_index]
    anime_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(anime.iloc[i[0]][1])
```

```python
recom('Kimi no Na wa.')
```

```python
recom('Naruto: Shippuuden')
```

```python
recom('One Piece: Cry Heart')
```

```python
recom('Shingeki no Kyojin Season 2')
```

```python
recom('Hunter x Hunter (2011)')
```
